```python
import pandas as pd

# Load the dataset into a DataFrame
restaurants_data = pd.read_csv('Dataset  intership.csv')

# Display the first few rows and inspect the columns
print(restaurants_data.head())
print(restaurants_data.columns)

# Calculate mean, median, and standard deviation for numerical columns
numerical_columns = ['Restaurant ID', 'Restaurant Name', 'City'] 

# Replace with actual numerical column names
statistics = restaurants_data[numerical_columns].describe()
print("\nstatistics:")
print(statistics)

# Count of restaurants per country code
country_counts = restaurants_data['Country Code'].value_counts()
print("Country Code Distribution:")
print(country_counts)

# Count of restaurants per city
city_counts = restaurants_data['City'].value_counts()
print("\nCity Distribution:")
print(city_counts)

# Count of restaurants per cuisine (assuming cuisines are in a list format within a column)
cuisine_counts = restaurants_data['Cuisines'].apply(lambda x: len(str(x).split(','))).value_counts()
print("\nCuisine Distribution:")
print(cuisine_counts)

# Top cuisines
top_cuisines = restaurants_data['Cuisines'].value_counts().head(10)
print("\nTop Cuisines:")
print(top_cuisines)

# Top cities with the highest number of restaurants
top_cities = restaurants_data['City'].value_counts().head(10)
print("\nTop Cities with Most Restaurants:")
print(top_cities)
```
