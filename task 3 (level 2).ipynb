```python
import pandas as pd

# Step 1: Import the dataset
df = pd.read_csv('Dataset  (2).csv')

# Display the first few rows of the dataset
print(df.head())

# Step 2: Extract additional features
# 2.1: Length of the restaurant name
df['Name_Length'] = df['Restaurant Name'].str.len()

# 2.2: Length of the address
df['Address_Length'] = df['Address'].str.len()

# Step 3: Create new features based on categorical variables
# 3.1: Has Table Booking 
df['Has_Table_Booking'] = df['Has Table booking'].apply(lambda x: 1 if x == 'Yes' else 0)

# 3.2: Has Online Delivery 
df['Has_Online_Delivery'] = df['Has Online delivery'].apply(lambda x: 1 if x == 'Yes' else 0)

# Step 4: Display the modified dataframe
print(df.head())

# Step 5: save the modified dataframe to a new CSV file 
df.to_csv('modified_dataset.csv', index=False)
```
